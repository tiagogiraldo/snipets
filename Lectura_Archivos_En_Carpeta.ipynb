### 2.  Lista de archivos en la carpeta

Esta opción lee los nombres de archivos en una lista y los almacena en una lista

```python
pathf = r'00_scimago_rank'
onlyfiles = [f for f in listdir(pathf) if isfile(join(pathf, f))]
```

### 3. Carga de archvos csv de una ruta

- Se leen los nombres de los archivos, pero esta vez, la lista contiene la ruta y para que el sistema los reconozca.
- Se abren con pandas y se anexan como una lista
- Se concatenan los archivos, para crear un df general con el contenido de toda la información.

```python
all_files = glob.glob(pathf + "/*.csv")

li = []

for filename in all_files:
    print(filename)
    df = pd.read_csv(filename, index_col=None, header=0, sep=';')
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True,sort=False)
```

### Filtro de journals Q1 y Q2 de los datos de scimago

- Se filtran las líneas que contengan Q1 y Q2
- Se copia un df del original
- Se extraen los valores únicos de journals


```python
df = frame[(frame['SJR Quartile']=='Q1') | (frame['SJR Quartile']=='Q2')]
```
